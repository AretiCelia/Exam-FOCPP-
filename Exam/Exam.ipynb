```python
#Task 5

# Creating an empty list to store the strings
strings = []

# Prompting the user to enter strings until an empty string is entered
while True:
    user_input = input("Enter a string (press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input.lower())

# Sort the strings alphabetically
strings.sort()


print("Strings in alphabetical order:")
for s in strings:
    print(s)
```
